```python
import tiktoken

# Caminho do arquivo de texto gerado pelo Whisper
arquivo = "transcricao_youtube.txt"

# Lê o conteúdo do arquivo
with open(arquivo, "r", encoding="utf-8") as f:
    texto = f.read()

# Contagem básica
num_caracteres = len(texto)
num_palavras = len(texto.split())

# Contagem de tokens (compatível com modelos GPT)
encoding = tiktoken.encoding_for_model("gpt-4o")  # ou "gpt-3.5-turbo", "gpt-4", etc.
num_tokens = len(encoding.encode(texto))

# Exibir resultados
print(f"📄 Arquivo: {arquivo}")
print(f"🔤 Caracteres: {num_caracteres:,}")
print(f"📝 Palavras: {num_palavras:,}")
print(f"🔢 Tokens (gpt-4o): {num_tokens:,}")
```

## Chunks com tiktoken e langchain

```python
import tiktoken
def chunk_text(text, model="gpt-4o", max_tokens=4000, overlap=200):
    encoding = tiktoken.encoding_for_model(model)
    tokens = encoding.encode(text)

    chunks = []
    start = 0

    while start < len(tokens):
        end = start + max_tokens
        chunk = tokens[start:end]
        chunks.append(encoding.decode(chunk))
        start += max_tokens - overlap

    print(f"Total de chunks gerados: {len(chunks)}")
    return chunks


#Exemplo
arquivo = 'transcricao_youtube.txt'
with open(arquivo, "r", encoding="utf-8") as f:
    texto = f.read()

chunks = chunk_text(texto, max_tokens=4000, overlap=200)
print(chunks[0][:500])  # Mostra início do primeiro chunk
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Define o tamanho do chunk em tokens aproximados
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=1000,
    chunk_overlap=200
)

# Lê o texto e divide
with open(arquivo, "r", encoding="utf-8") as f:
    texto = f.read()

docs = text_splitter.create_documents([texto])

print(f"Total de chunks: {len(docs)}")
print(docs[1].page_content[:500])  # Mostra o início do segundo chunk
```
