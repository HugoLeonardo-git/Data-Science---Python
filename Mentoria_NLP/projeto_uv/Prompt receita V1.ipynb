# Transcrição da receita

```python
TEXTO_TRANSCRICAO = pd.read_csv('data/transcricao_receita.txt', delimiter=";")
```

# Prompt de receita

```python
prompt_receita = f"""Você é um assistente culinário especializado em extrair informações de receitas a partir de transcrições de fala.

A seguir está a transcrição completa de um vídeo de receita. 
Sua tarefa é transformar esse texto desestruturado em um formato organizado e padronizado.

=== INSTRUÇÕES ===
1. Identifique a classificação do prato: "Entrada", "Prato principal" ou "Sobremesa".
2. Extraia todos os ingredientes mencionados, mesmo que apareçam no meio da explicação.
3. Extraia o modo de preparo em passos claros e objetivos.
4. Detecte o tempo estimado total da receita, considerando:
    • preparo
    • cocção
    • descanso (se houver)
5. Se algo estiver ausente ou não for mencionado, responda "Não informado".

=== FORMATO DA RESPOSTA (OBRIGATÓRIO) ===
Classificação: <tipo_do_prato>

Ingredientes:
- item 1
- item 2
- ...

Modo de Preparo:
1. passo 1
2. passo 2
3. ...

Tempo estimado: <minutos ou 'Não informado'>

=== TRANSCRIÇÃO A SEGUIR ===
{TEXTO_TRANSCRICAO}
"""
```

```python
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": prompt_receita}]
)
resultado = response.choices[0].message.content
print(resultado)
```
