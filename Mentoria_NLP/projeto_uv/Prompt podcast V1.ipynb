# Texto do podcast

```python
TEXTO_TRANSCRICAO = pd.read_csv('data/transcricao_podcast.txt', delimiter=";")
```

# Prompt de podcast

```python
prompt = f"""Você é um analista especializado na avaliação de conteúdos de podcast. 
A seguir está a transcrição completa de um episódio. 
Sua tarefa é analisar o conteúdo e produzir um resumo estruturado e claro.

=== INSTRUÇÕES ===
Com base na transcrição, identifique e descreva:

1. **Expertise do convidado**  
   - Formação, área de atuação, principais projetos, temas dominados.

2. **Expertise do host (se houver)**  
   - Perfil profissional, área de conhecimento, estilo de condução.

3. **Participantes**  
   - Liste todos que aparecem na conversa e seus papéis.

4. **Tópicos principais discutidos**  
   - Separe por blocos temáticos, tópicos macro e tópicos micro.  
   - Para cada tópico, indique brevemente o que foi argumentado.

5. **Referências**  
   - Pesquisas citadas  
   - Dados mencionados  
   - Autores, livros, papers ou estudos  
   - Links ou instituições mencionadas

6. **Pontos de destaque do episódio** (adicional recomendado)
   - Insights importantes  
   - Ideias inovadoras  
   - Controvérsias ou debates interessantes

7. **Conclusões e mensagens finais** (adicional recomendado)
   - Sínteses  
   - Recomendações  
   - Lições principais

8. **Resumo geral em um parágrafo**  
   - Objetivo, essência da conversa e principais aprendizados.

=== FORMATO DA RESPOSTA ===

Expertise do convidado:
<texto>

Expertise do host:
<texto ou "Não informado">

Participantes:
- nome (papel)
- nome (papel)

Tópicos discutidos:
1. <tópico> — <descrição>
2. <tópico> — <descrição>
...

Referências:
- <referência 1>
- <referência 2>
...

Pontos de destaque:
- <ponto 1>
- <ponto 2>
...

Conclusões:
<texto>

Resumo geral:
<parágrafo único>

=== TRANSCRIÇÃO A SEGUIR ===
{TEXTO_TRANSCRICAO}
"""
```

```python
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": prompt}]
)
resultado = response.choices[0].message.content
print(resultado)
```
